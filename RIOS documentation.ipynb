Here - we are geting unique index for the numpar by using Unique function.

```python
length_numpar = len((df.numpar).unique())
```

In the 1st line: We are extarcting the required column such as ('numpar', 'WaterDepth', 'depth) from data frame'df' and assigning it to the "onlyLastRow".

In the second line we are intilizing the counter assigning by the variale "count_num800Above" in order to count number of the larvae above 800 meter depth.

In the 3rd line we are using 'for' loop in the range of (length of numpa)r) which would be the (479).

In the 4th line we are geting the last row of the every numpar(numer of larvae) from the dataframe'df' using 
(. tail(1)) function.we have assigned it as numpar_single.

In the 5th line we are using if-condition to check the last row if it is above 800 meter.

In the 6th line ,we are counting every last row which meets the given if condition.

In the 7th line we are creating pandas series where we are going to store the index,waterdepth,depth of the last row of every numpar.

In the 8th line we are using append function to append every selected columns from the last row of numpar together in order to form the table.

```python

onlyLastRow = pd.DataFrame(columns=['numpar', 'WaterDepth', 'depth'])
count_num800Above = 0
for i in range (length_numpar): 
    numpar_single = df.loc[df['numpar'] == i].tail(1)
    if ( float(numpar_single.WaterDepth) < 2000):
        count_num800Above += 1
    s = pd.Series([int(i),float(numpar_single.WaterDepth), float(numpar_single.depth)],index=['numpar','WaterDepth', 'depth'])
    onlyLastRow = onlyLastRow.append(s,ignore_index=True)
    
    
    
```
